# 5. Bivariate Analysis

### Correlation Matrix

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

The correlation matrix shows some interesting relationships. For example, PM2.5 and PM10 are moderately correlated. There are also some weak correlations between pollutants and weather conditions.

### Scatter Plots

```python
sns.scatterplot(data=df, x='Temperature', y='PM2.5')
plt.title('Temperature vs. PM2.5')
plt.show()

sns.scatterplot(data=df, x='Humidity', y='PM2.5')
plt.title('Humidity vs. PM2.5')
plt.show()

sns.scatterplot(data=df, x='Wind Speed', y='PM2.5')
plt.title('Wind Speed vs. PM2.5')
plt.show()
```

# 6. Multivariate Analysis

### Pair Plot

```python
sns.pairplot(df, vars=['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3'])
plt.show()
```

The pair plot helps to visualize the relationships between all pairs of pollutants. We can see the correlation between PM2.5 and PM10 more clearly here.

# 7. Geospatial Analysis

```python
import folium
from geopy.geocoders import Nominatim

# Get city coordinates
geolocator = Nominatim(user_agent='air_quality_eda')
city_coords = {}
for city in df['City'].unique():
    try:
        location = geolocator.geocode(city)
        if location:
            city_coords[city] = (location.latitude, location.longitude)
    except:
        pass

# Create map
m = folium.Map(location=[20, 0], zoom_start=2)

# Add markers
for city, coords in city_coords.items():
    avg_pm25 = df[df['City'] == city]['PM2.5'].mean()
    folium.Marker(
        location=coords,
        popup=f'{city}: {avg_pm25:.2f} PM2.5',
        tooltip=city
    ).add_to(m)

m
```

The map above shows the average PM2.5 levels for each city. The markers are placed on the map according to the city's coordinates. The popup shows the city name and the average PM2.5 level.

# 8. Time Series Analysis

```python
df.set_index('Date', inplace=True)
df['PM2.5'].plot(figsize=(15, 7), title='PM2.5 Over Time')
plt.ylabel('PM2.5')
plt.show()
```

```python
df.resample('M')['PM2.5'].mean().plot(figsize=(15, 7), title='Monthly Average PM2.5')
plt.ylabel('PM2.5')
plt.show()
```

The time series plots show the trend of PM2.5 over time. The first plot shows the daily PM2.5 levels, and the second plot shows the monthly average PM2.5 levels. This can help to identify any seasonality or long-term trends in the data.

# 10. Country-based Analysis

```python
country_mean = df.groupby('Country')[['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']].mean().sort_values(by='PM2.5', ascending=False)
country_mean
```

### Bar plots for each pollutant by country

```python
for col in country_mean.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x=country_mean.index, y=country_mean[col])
    plt.title(f'Average {col} by Country')
    plt.xticks(rotation=90)
    plt.show()
```

### Box plot of PM2.5 by Country

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='Country', y='PM2.5')
plt.xticks(rotation=90)
plt.title('Distribution of PM2.5 by Country')
plt.show()
```
